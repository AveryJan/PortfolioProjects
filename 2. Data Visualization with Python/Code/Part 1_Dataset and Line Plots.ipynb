# Preparing Data <a id="2"></a>

### 1. Remove the columns that are not needed for data visualization.


```python
# Note: In pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head()
```

### 2. Rename some columns to meaningful names.


```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

### 3. Add a new column 'Total' that sums up the total immigrants by country over the entire period 1980 - 2013.


```python
# Define a new column "Total.'

df_can['Total'] = df_can.sum(axis=1)

# Check if the new column is added to the dataframe. 
df_can.head()
```

### 4. Check if there is any null object in the dataset. 


```python
# Look for any null objects in the dataset because when creating visualizations, any NaN, i.e. not a number, values 
# will default to 0 .

df_can.isnull().sum()
```

#### Comment
No missing values in the Canada Immigration Dataset. 

### 5. Get the statistical summary of the numerical columns. 

```python
# Get the statistical summary. 
df_can.describe()
```

### 6. Set the country name as index to facilitate the selection of rows of data by country.


```python
# Set the country name as the index. 
df_can.set_index('Country', inplace=True)

# Confirm that the country names are now the indices of the dataframe. 
df_can.head(3)
```

```python
# Rmove the name of the index
df_can.index.name = None
```

### 7. Convert column labels to 'string' to avoid confusion introduced by having an integer as the column name.


```python
# Convert all column labels to 'string' type. 
df_can.columns = list(map(str, df_can.columns))

# Confirm that all column labels are now 'string' type. 
print(df_can.columns)
```

### 8. Create a list of the names of year columns (1980-2013, in type 'string') to be used for creating visualizations

```python
# Since the column names of the year columns were converted to string, a variable that holds a list of the names of all 'year' columns' 
# is declared so that the full range of years can be called upon when creating visualizations. 

years = list(map(str, range(1980, 2014)))
years
```

## 1. A case study on Haiti

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and about three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti, resulting in a spike in immigration from Haiti to Canada. 

```python
# Goal: Plot a line graph of immigration from Haiti.

# Step 1: Extract the data series for Haiti.
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

Next, we will plot a line plot by appending `.plot()` to the `haiti` dataframe.


```python
# Step 2: Change the type of the index values (the years) to integer from string for plotting.
haiti.index = haiti.index.map(int)

# Step 3: Create the line plot.
haiti.plot(kind='line')

# Step 4: Add a titel and axes labels. 
plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# Step 5: Annotate the 2010 Earthquake. 
plt.text(2000, 6000, '2010 Earthquake') # Note: the syntax is plt.text(x, y, label for the annotation)

# Step 6: Display the plot. 
plt.show() 
```

## 2. Comparison of immigration from China and India

```python
# Goal: Compare the number of immigrants from India and China from 1980 to 2013.

# Step 1: Get the data set for China and India, and display the resulting dataframe.
df_CI = df_can.loc[['India', 'China'], years]
df_CI
```

```python
# Step 2: Transpose the dataframe.

## Note: Pandas plots the indices on the x-axis and the columns as individual lines on the y-axis. 
## Since df_CI is a dataframe with the country as the index and years as the columns, it is necessary to transpose the dataframe 
## using transpose() method to swap the row and columns.

df_CI = df_CI.transpose()
df_CI.head()
```

```python
# Step 3: Change the index values of df_CI to type integer for plotting.
df_CI.index = df_CI.index.map(int) 

# Step 4: Create the line plot.
df_CI.plot(kind='line')

# Step 5: Add a title and axes labels.
plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# Step 6: Display the plot. 
plt.show()
```

### Comment
From the above plot, we can observe that the China and India have very similar immigration trends through the years.

## 3. The trend of top 5 countries that contributed the most to immigration to Canada.


```python
    #Step 1: Get the dataset. Recall that we created a Total column that calculates cumulative immigration by country. 
    #We will sort on this column to get our top 5 countries using pandas sort_values() method.
    
    inplace = True # paramemter saves the changes to the original df_can dataframe
    df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

    # get the top 5 entries
    df_top5 = df_can.head(5)

    # transpose the dataframe
    df_top5 = df_top5[years].transpose() 

    print(df_top5)


    #Step 2: Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
    df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
    df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size



    plt.title('Immigration Trend of Top 5 Countries')
    plt.ylabel('Number of Immigrants')
    plt.xlabel('Years')
```

### Comment

As the plot above shows, plotting five lines on a single graph starts to make the interpretation of the data more difficult. However, for the most part, the plot indicates that United Kingdom of Great Britain and Northern Ireland bucks the trend. Its immigration to Canada decreased dramatically between 1995 and 2000 as opposed to an uptrend displayed by the other four countries.
